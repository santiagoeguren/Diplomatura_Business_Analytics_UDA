**Ejercicio Nº3**: El gerente de una tienda de estéreos y equipos de estéreo quiere determinar la relación entre el número de comerciales de televisión transmitidos el fin de semana y las ventas en la tienda durante la semana siguiente.  Para ello se registraron 10 observaciones, una para cada semana. Las ventas están medido en dólares. Suponga Varianza poblacional finita.





```
numero_de_comerciales=c(2,5,1,3,4,1,5,3,4,2)
ventas=c(5000,5700,4100,5400,5400,3800,6300,4800,5900,4600)
```

```
#Gráficar

plot(numero_de_comerciales,ventas)
```

```
#Calcular la correlación

cor(numero_de_comerciales,ventas)
```
